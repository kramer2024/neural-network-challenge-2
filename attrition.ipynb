## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
display(y_df.head())
display(y_df.dtypes)
```

```python
# Create a list of at least 10 column names to use as X data
x_columns = ['Education', 'Age', 'DistanceFromHome', 'JobSatisfaction', 'YearsInCurrentRole', 'StockOptionLevel', 'WorkLifeBalance', 'YearsAtCompany', 'YearsSinceLastPromotion', 'NumCompaniesWorked']

# Create X_df using your selected columns
X_df = attrition_df[x_columns]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state= 14)
```

```python
# display contents of X and y dataframes for reference
display(X_train.head())
display(y_train.head())
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train = X_train.astype('int64')
X_test = X_test.astype('int64')
```

```python
# display elements of data
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
display(X_train.dtypes)
display(X_test.dtypes)
display(y_train.dtypes)
display(y_test.dtypes)
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# import required library to run encoders properly within cell
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create a OneHotEncoder for the Department column
hot_encoder_dept = OneHotEncoder(sparse_output = False)

# Fit the encoder to the training data
hot_encoder_dept.fit(y_train['Department'].values.reshape(-1, 1))

# Create two new variables by applying the encoder
# to the training and testing data
y_train_department_encoded = hot_encoder_dept.transform(y_train['Department'].values.reshape(-1, 1))
y_test_department_encoded = hot_encoder_dept.transform(y_test['Department'].values.reshape(-1, 1))

# display encoded data
y_test_department_encoded
```

```python
# Create a OneHotEncoder for the Attrition column
hot_encoder_att = OneHotEncoder(sparse_output = False)

# Fit the encoder to the training data
hot_encoder_att.fit(y_train['Attrition'].values.reshape(-1, 1))

# Create two new variables by applying the encoder
# to the training and testing data
y_train_attrition_encoded = hot_encoder_att.transform(y_train['Attrition'].values.reshape(-1, 1))
y_test_attrition_encoded = hot_encoder_att.transform(y_test['Attrition'].values.reshape(-1, 1))

# display encoded data
y_test_attrition_encoded[:10]
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_layers = X_train_scaled.shape[1]

# Create the input layer
inputs = layers.Input(shape = (input_layers,))

# Create at least two shared layers
shared_layer_1 = layers.Dense(64, activation= 'relu', name = 'shared1')(inputs)
shared_layer_2 = layers.Dense(128, activation= 'relu', name = 'shared2')(shared_layer_1)
```

```python
# Create a branch for Department with a hidden layer and an output layer

# Create the hidden layer
hidden_dept = layers.Dense(32, activation= 'relu', name = 'department_hidden')(shared_layer_2)

# Create the output layer
output_dept = layers.Dense(3, activation= 'softmax', name = 'department_output')(hidden_dept)
```

```python
# Create a branch for Attrition with a hidden layer and an output layer

# Create the hidden layer
hidden_attr = layers.Dense(32, activation= 'relu', name = 'attrition_hidden')(shared_layer_2)

# Create the output layer
output_attr = layers.Dense(2, activation= 'sigmoid', name = 'attrition_output')(hidden_attr)
```

```python
# Create the model
att_model = Model(inputs = inputs, outputs = [output_dept, output_attr])

# Compile the model
att_model.compile(optimizer = 'adam',
                  loss = {'department_output': 'categorical_crossentropy', 'attrition_output': 'binary_crossentropy'},
                  metrics = {'department_output': 'accuracy', 'attrition_output': 'accuracy'})

# Summarize the model
att_model.summary()
```

```python
# Train the model
att_model.fit(X_train_scaled,
              {'department_output': y_train_department_encoded, 'attrition_output': y_train_attrition_encoded},
              epochs = 100,
              batch_size = 32,
              verbose = 2)
```

```python
# Evaluate the model with the testing data
test_results = att_model.evaluate(X_test_scaled,
                                  {'department_output': y_test_department_encoded,
                                   'attrition_output': y_test_attrition_encoded})
test_results
```

```python
# Print the accuracy for both department and attrition
print(f'Department predictions accuracy: {test_results[3]}')
print(f'Attrition predictions accuracy: {test_results[4]}')
```
